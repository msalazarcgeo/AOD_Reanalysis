```python
import rasterio
import os
```

```python

def set_band_names_gee_PBLH(file_path):
    """Rename the bands in the file """
    dst_m= rasterio.open(file_path,'r+',  driver =  'GTiff')
    if dst_m.count == 9:
        band_names = [
              "dewpoint_temperature_2m",
              "total_evaporation",
              "temperature_2m",
              "total_precipitation_hourly",
              "total_precipitation",
              "relative_humidity",
              "Optical_Depth_047",
              "AOD_Uncertainty",
              "PBLH"
        ]
    elif dst_m.count == 7:
        band_names = [
              "dewpoint_temperature_2m",
              "total_evaporation",
              "temperature_2m",
              "total_precipitation_hourly",
              "total_precipitation",
              "relative_humidity",
              "PBLH"
        ]
    for num, band in enumerate(band_names ):
        dst_m.set_band_description(num+1, band)
    dst_m.close()
```

```python
path = './datos/raster/ZMG_redo/middle_PBLH_ZMG/all_bands/'
years =  os.listdir(path)
years = [path+ year for year in years]
for year_pat in years:
    list_file = os.listdir(year_pat)
    
    for file_pa in  list_file:
        print(year_pat+'/'+file_pa)
        set_band_names_gee_PBLH(year_pat+'/'+file_pa)
```

```python
def count_bands(filepath):
    """Print the number of band"""
    
    dst_m= rasterio.open(filepath,'r',  driver =  'GTiff')
    band_c = dst_m.count
    dst_m.close()
    return(band_c)
    
```
